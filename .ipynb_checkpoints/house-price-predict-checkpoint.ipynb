### Assumptions on Multiple regression
Before proceeding to get's hand's dirty one should know what are the prerequisites or assumptions regarding Multiple regression. I am iterating the assumption as below:
* There must be a linear relationship between the outcome variable and the independent variables.  
    We will use Scatterplots can show whether there is a linear or curvilinear relationship.
* Multivariate Normality–Multiple regression assumes that the residuals are normally distributed.
    We will use Q-Q plot with scatter plots to verify
* No Multicollinearity—Multiple regression assumes that the independent variables are not highly correlated with each other.  
    We will use corrmat to review Multicollinearity.
* Homoscedasticity–This assumption states that the variance of error terms are similar across the values of the independent variables.  A plot of standardized residuals versus predicted values can show whether points are equally distributed across all values of the independent variables.

Lets develop a function to impute the imssing variables. We will use RandomForest to impute the missing values for LotFrontage.

```python
def imputeMissing(df, idx, predictors):
    tempdf = df[df[outcome].isnull()].index
    tempndf = df[df[idx].notnull()].index
    model = RandomForestRegressor(n_estimators=100)
    model.fit(df.loc[tempndf, predictors], df.loc[tempndf, [outcome]])
    predict = model.predict(df.loc[tempndf, predictors])
    j=0;
    for i in tempdf:
        df.loc[i,outcome] = predict[j]
        j=j+1
    return(df)
```

We will use the other lot features to impute the missing values for LotFrontage variable.

```python
predictors=['LotArea','Street','LotShape','LandContour','LotConfig']
outcome='LotFrontage'

## Encode the variable to numeric 
var_mod = ['Street','LotShape','LandContour','LotConfig']
le = LabelEncoder()
for i in var_mod:
    fullData[i] = le.fit_transform(fullData[i].astype(str))

fullData = imputeMissing(fullData,outcome,predictors)
fullData['LotFrontage'].isnull().sum()
```

## Design and building Predictive Model

As we have completed most tedious task exploratory analysis and data cleaning, next we will build our first predictive model and further tune it using corss validation and viewing results. While building our first model we will skip some columns which were not fixed during imputation and not relevent.

```python
## Make a list of columns to skip by our model
skipedCols = ['SalePrice', 'PoolQC','MiscFeature','Alley','Id','Fence','FireplaceQu','dataType']
## Select all features expect skiped columns
predictors = [x for x in trainData.columns if x not in skipedCols]


## Encode the predictors which are categorical
le = LabelEncoder()
for i in predictors:
    trainData[i] = le.fit_transform(trainData[i].astype(str))
    
seed = 7
model1 = GradientBoostingRegressor(random_state=10)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(model1, trainData[predictors], trainData['SalePrice'],scoring='r2', cv=kfold)

### View the first model performance metrics
print("CV Score")
print("Mean : {}".format(round(np.mean(results),4)))
print("Standard Deviation : {}".format(round(np.std(results),4)))
print("Min : {}".format(round(np.min(results),4)))
print("Max : {}".format(round(np.max(results),4)))
print("\n")
print("Ok! our CV mean score {} is good but we can improve it further".format(round(np.mean(results),4)))
print("First lets also visualize which features contribute to our model and which are safe to ignore by plotting them.")
```

```python
model1.fit(trainData[predictors], trainData['SalePrice'])
feat_imp = pd.Series(model1.feature_importances_, predictors).sort_values(ascending=False)
feat_imp = feat_imp[feat_imp > 0.0]
plt.figure(figsize=(16,6))
feat_imp.plot(kind='bar', title='Feature Importances')
plt.ylabel('Feature Importance Score')
```

### Tuning Our Model

Let's kickstart tuning our model for parameter __n_estimators__. You can refer to link : [scikit](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html) for more details on parameters. Before tuning parameters one by one first let take some dafult values and tune n_estimators.

```python
paramList1 = {'n_estimators':range(20,161,10)}
model2 = GridSearchCV(estimator = 
                        GradientBoostingRegressor(learning_rate=0.1, min_samples_split=500,min_samples_leaf=50,max_depth=8,
                                                  max_features='sqrt',subsample=0.8,random_state=10), 
                        param_grid = paramList1, scoring='r2',n_jobs=4,iid=False, cv=5)
model2.fit(trainData[predictors], trainData['SalePrice'])
print("Optimal value for n_estimators :{}".format(model2.best_params_['n_estimators']))
print("Model best Score: {}".format(round(model2.best_score_,4)))
print('R squared: {}'.format(round(model2.score(trainData[predictors], trainData['SalePrice']),4)))
```

As with our crossvalidation GridSearchCV iteration we got best estimators for 'n_estimators'. Next lets tune parameters max_depth and min_samples_split using 'n_estimators'. Now further tuning I’ll test max_depth values of 5 to 16 with interval of 2 and min_samples_split from 100 to 1001 in steps of 200. The values that I have choosed are based on my own intution. You can set wider ranges as well and then perform multiple iterations for smaller ranges.

```python
paramList2 = {'max_depth':range(5,16,2), 'min_samples_split':range(100,1001,200)}
model3 = GridSearchCV(estimator = 
                      GradientBoostingRegressor(learning_rate=0.1, n_estimators=model2.best_params_['n_estimators'],
                                                max_features='sqrt', subsample=0.8, random_state=10), 
                      param_grid = paramList2, scoring='r2',n_jobs=4,iid=False, cv=5)
model3.fit(trainData[predictors], trainData['SalePrice'])
print("Optimal value for max_depth : {}".format(model3.best_params_['max_depth']))
print("Optimal value for min_samples_split : {}".format(model3.best_params_['min_samples_split']))
print("Model best Score: {}".format(round(model3.best_score_,4)))
print('R squared: {}'.format(round(model3.score(trainData[predictors], trainData['SalePrice']),4)))
```

Our corssvalidation reported tuned parameters for max_depth and min_samples_split as above and lets stick with it and tune min_samples_leaf parameter.

```python
paramList3 = {'min_samples_leaf':range(5,61,5)}
model4 = GridSearchCV(estimator = 
                        GradientBoostingRegressor(learning_rate=0.1,n_estimators= model2.best_params_['n_estimators'],
                                                  max_depth=model3.best_params_['max_depth'],
                                                  min_samples_split=model3.best_params_['min_samples_split'],
                                                  max_features='sqrt',
                                                  subsample=0.8, random_state=10), 
                        param_grid = paramList3, scoring='r2',n_jobs=4,iid=False, cv=5)
model4.fit(trainData[predictors], trainData['SalePrice'])
print("Optimal value for min_samples_leaf : {}".format(model4.best_params_['min_samples_leaf']))
print("Model best Score: {}".format(round(model4.best_score_,4)))
print('R squared: {}'.format(round(model4.score(trainData[predictors], trainData['SalePrice']),4)))
```

So our tuned parameters for min_samples_leaf is 10 and lets stick with it and tune max_features parameter.

```python
paramList4 = {'max_features':range(20,70,2)}
model5 = GridSearchCV(estimator = 
                      GradientBoostingRegressor(
                          learning_rate=0.1, n_estimators= model2.best_params_['n_estimators'],
                          max_depth=model3.best_params_['max_depth'],
                          min_samples_split=model3.best_params_['min_samples_split'], 
                          min_samples_leaf=model4.best_params_['min_samples_leaf'],
                          subsample=0.8, random_state=10),
                      param_grid = paramList4, scoring='r2',n_jobs=4,iid=False, cv=5)
model5.fit(trainData[predictors], trainData['SalePrice'])
print("Optimal value for max_features : {}".format(model5.best_params_['max_features']))
print("Model best Score: {}".format(round(model5.best_score_,4)))
print('R squared: {}'.format(round(model5.score(trainData[predictors], trainData['SalePrice']),4)))
```

Now we have tuned our model at a level, so lets build our final model which we will use for prediction on test dataset.

```python
estimator = GradientBoostingRegressor(
    learning_rate=0.1, n_estimators=model2.best_params_['n_estimators'],
    max_depth=model3.best_params_['max_depth'],
    min_samples_split=model3.best_params_['min_samples_split'], 
    min_samples_leaf=model4.best_params_['min_samples_leaf'],
    max_features=model5.best_params_['max_features'],
    subsample=0.8, random_state=10)

estimator.fit(trainData[predictors], trainData['SalePrice'])
print('Gradient Boosting Final R squared": {}'.format(round(estimator.score(trainData[predictors], trainData['SalePrice']),4)))
```
